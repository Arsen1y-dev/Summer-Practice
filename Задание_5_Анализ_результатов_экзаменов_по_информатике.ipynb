Чтобы результаты были корректными, следует учитывать только те районы, в которых достаточно учеников для значимого анализа. Так же это поможет при дальнейшем определении лучшей и худшей школы, т.к. если это не учитывать, то лучшей школой окажется та, которой таковой не является.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Получаем данные по районам с более чем 100 учениками
area_counts = oge1['Район'].value_counts()
major_areas = area_counts[area_counts >= 100].index
filtered_data = oge1[oge1['Район'].isin(major_areas)]

# Рассчитываем среднюю оценку по районам
average_scores_by_area = filtered_data.groupby('Район')['Оцека по 5'].mean().sort_values()

# Находим районы с наивысшими и наименьшими средними оценками
best_area = average_scores_by_area.idxmax()
best_score = average_scores_by_area.max()

worst_area = average_scores_by_area.idxmin()
worst_score = average_scores_by_area.min()

plt.figure(figsize=(12, 8))

# Устанавливаем стиль визуализации
sns.set(style='whitegrid')

# Строим горизонтальный график с улучшениями
bar_plot = sns.barplot(
    x=average_scores_by_area,
    y=average_scores_by_area.index,
    palette='viridis',  # Используем палитру 'viridis' для лучшей визуализации
    ci=None
)

# Добавляем значения на график для каждой полосы
for p in bar_plot.patches:
    width = p.get_width()
    bar_plot.text(width + 0.02, p.get_y() + p.get_height()/2,
                  f'{width:.2f}', ha='left', va='center', fontsize=11, color='black')

# Добавляем аннотации для лучшего и худшего района
bar_plot.text(best_score + 0.05, average_scores_by_area.index.get_loc(best_area) + 0.2,
              f'Лучший район: {best_area}\nСредний балл: {best_score:.2f}',
              ha='left', va='center', fontsize=12, color='green', weight='bold')

bar_plot.text(worst_score + 0.05, average_scores_by_area.index.get_loc(worst_area) - 0.2,
              f'Худший район: {worst_area}\nСредний балл: {worst_score:.2f}',
              ha='left', va='center', fontsize=12, color='red', weight='bold')

# Настраиваем заголовок и метки
plt.title('Средняя оценка по районам с количеством учеников более 100', fontsize=18, fontweight='bold')
plt.xlabel('Средняя оценка', fontsize=14)
plt.ylabel('Район', fontsize=14)

# Убираем ненужные элементы и настраиваем оси
sns.despine(left=True, bottom=True)
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Оптимизируем размещение элементов
plt.tight_layout()

# Отображаем график
plt.show()

# Выводим информацию о лучшем и худшем районе в консоль
print(f'Район с наивысшим средним баллом: {best_area} (Средний балл: {best_score:.2f})')
print(f'Район с наименьшим средним баллом: {worst_area} (Средний балл: {worst_score:.2f})')
```

Лучшая школа: МАОУ "Гимназия № 3" \
Худшая школа: МБОУ - СОШ с. Новосельское


```python
# Фильтрация данных для гимназии №3
school_data = oge1[oge1['Краткое наименование ОО'] == 'МАОУ "Гимназия № 3"']

# Подсчитываем количество учеников, распределение оценок
grades_distribution = school_data['Оцека по 5'].value_counts().sort_index()
total_students = grades_distribution.sum()

# Выводим количество учеников
print(f'Всего учеников: {total_students}')

# Определяем метки оценок на основе уникальных значений
grade_labels = [f'Оценка {grade}' for grade in grades_distribution.index]

# Устанавливаем новый стиль через seaborn
sns.set_theme(style="darkgrid")

# Создаем фигуру и ось для визуализации
fig, ax = plt.subplots(figsize=(8, 8))

# Добавляем круговую диаграмму с новыми цветами и измененными параметрами
color_scheme = sns.color_palette("Set2")

# Построение круговой диаграммы с увеличенными долями и измененными параметрами
slices, labels, percents = ax.pie(
    grades_distribution, labels=grade_labels, colors=color_scheme, autopct='%.2f%%',
    startangle=90, explode=[0.05] * len(grades_distribution), wedgeprops={'linewidth': 1.5, 'edgecolor': 'black'}
)

# Улучшение текста процентов и меток
for text in percents:
    text.set_color('navy')
    text.set_fontsize(11)
    text.set_fontweight('bold')

# Заголовок и аннотация с общим количеством учеников
ax.set_title('Распределение оценок ОГЭ \nГимназии № 3', fontsize=15, weight='bold', color='darkred')
ax.annotate(f'Всего учеников: {total_students}', xy=(0, -1.2), fontsize=12, ha='center', color='black')

# Делаем диаграмму круговой
ax.axis('equal')

# Окончательное форматирование и вывод
plt.tight_layout()
plt.show()
```

Мы видим, что наилучшие результаты ОГЭ показали гимназии. За ними следуют лицеи, а затем средние общеобразовательные школы с углубленным изучением отдельных предметов. Давайте рассмотрим эти категории более подробно.


```python
# Исходные данные
lic = oge.loc[oge['Вид ОО'] == 'Лицей']
grade_l = lic['Оцека по 5'].mean()

gim = oge.loc[oge['Вид ОО'] == 'Гимназия']
grade_g = gim['Оцека по 5'].mean()

sred = oge.loc[oge['Вид ОО'] == 'Средняя общеобразовательная школа с углубленным изучением отдельных предметов']
grade_s = sred['Оцека по 5'].mean()

data = {
    'Тип образовательного учреждения': ['Лицей', 'Гимназия', 'СОШ с углубленным изучением отдельных предметов'],
    'Средняя оценка': [grade_l, grade_g, grade_s]
}

df = pd.DataFrame(data)

plt.figure(figsize=(10, 6))
sns.set_style("whitegrid")

sns.barplot(x='Тип образовательного учреждения', y='Средняя оценка', data=df, palette='coolwarm', edgecolor='black')

plt.title('Средние оценки по типам образовательных учреждений', fontsize=16, fontweight='bold')
plt.xlabel('Тип образовательного учреждения', fontsize=12)
plt.ylabel('Средняя оценка по 5-балльной шкале', fontsize=12)

plt.xticks(rotation=45, ha='right', fontsize=10)
plt.yticks(fontsize=10)
sns.despine(left=True, bottom=True)

plt.tight_layout()

plt.show()
```

```python
# Функция для преобразования символов из строки
def convert_symbols(sequence):
    return [1 if char == '+' else 0 for char in sequence]

# Применение преобразований к столбцу 'Часть В' для вычисления баллов
oge1['Баллы за часть В'] = oge1['Часть В'].map(lambda seq: sum(convert_symbols(seq)))

# Преобразование символов в числовые значения и расчет сумм
oge1['сумма символов часть В'] = oge1['Часть В'].apply(lambda seq: sum(convert_symbols(seq)))
oge1['сумма символов часть C'] = oge1['Первичный балл'] - oge1['сумма символов часть В']

# Рассчитываем средние значения по районам
sr_b = oge1.groupby('Район')['сумма символов часть В'].mean()
sr_c = oge1.groupby('Район')['сумма символов часть C'].mean()

# Объединяем данные для построения графика
comparison_df = pd.DataFrame({
    'Среднее значение части В': sr_b,
    'Среднее значение части C': sr_c
}).reset_index()

# Настройка визуализации
sns.set(style='whitegrid')

plt.figure(figsize=(10, 6))

# Построение графика рассеяния
sns.scatterplot(x='Среднее значение части В', y='Среднее значение части C', data=comparison_df, color='teal', s=100)

# Добавляем линию регрессии
sns.regplot(x='Среднее значение части В', y='Среднее значение части C', data=comparison_df, scatter=False, color='red', line_kws={'linewidth': 2})

plt.title('Среднее значение баллов за часть В vs. часть C по районам', fontsize=16, fontweight='bold')
plt.xlabel('Среднее значение части В', fontsize=12)
plt.ylabel('Среднее значение части C', fontsize=12)
plt.grid(True)

# Отображаем график
plt.tight_layout()
plt.show()
```
