###Анализ результатов по заданиям

Процент успешного решения заданий ЕГЭ

```python
 ege_task_success_rate = ege.apply(lambda row: row[f'Task{row.index}'] / ege['Балл'].sum() * 100, axis=1)
ege_task_success_rate.sort_values(ascending=False, inplace=True)
print(ege_task_success_rate)
```

Процент успешного решения заданий ОГЭ

```python
oge_task_success_rate = oge.apply(lambda row: row[f'{row.index}_Часть В'] / oge['% Оценка'].sum() * 100, axis=1)
oge_task_success_rate.sort_values(ascending=False, inplace=True)
print(oge_task_success_rate)
```

```python
sns.barplot(x=oge_task_success_rate.index, y=oge_task_success_rate.values)
plt.title('Процент успешного решения заданий ОГЭ по информатике')
plt.xlabel('Номер задания')
plt.ylabel('Процент успешного решения')
plt.xticks(rotation=90)
plt.show()
```

###"Профили" отличников, хорошистов, двоечников

"Профиль" отличника по ЕГЭ

```python
ege_top_performers = ege[ege['Балл'] >= 85]
ege_top_performers_task_success_rate = ege_top_performers.apply(lambda row: row[f'Task{row.index}'] / ege_top_performers['Балл'].sum() * 100, axis=1)
ege_top_performers_task_success_rate.sort_values(ascending=False, inplace=True)
print('Профиль отличника по ЕГЭ:')
print(ege_top_performers_task_success_rate)
```

"Профиль" хорошиста по ЕГЭ

```python
ege_good_performers = ege[(ege['Балл'] >= 65) & (ege['Балл'] < 85)]
ege_good_performers_task_success_rate = ege_good_performers.apply(lambda row: row[f'Task{row.index}'] / ege_good_performers['Балл'].sum() * 100, axis=1)
ege_good_performers_task_success_rate.sort_values(ascending=False, inplace=True)
print('Профиль хорошиста по ЕГЭ:')
print(ege_good_performers_task_success_rate)
```

"Профиль" двоечника по ЕГЭ

```python
ege_bad_performers = ege[ege['Балл'] < 40]
ege_bad_performers_task_success_rate = ege_bad_performers.apply(lambda row: row[f'Task{row.index}'] / ege_bad_performers['Балл'].sum() * 100, axis=1)
ege_bad_performers_task_success_rate.sort_values(ascending=False, inplace=True)
print('Профиль двоечника по ЕГЭ:')
print(ege_bad_performers_task_success_rate)
```
