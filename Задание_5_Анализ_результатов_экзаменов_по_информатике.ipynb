###Анализ результатов по видам школ

Средний балл ЕГЭ по видам школ

```python
sns.set_style('whitegrid')

grade = oge.groupby('Вид ОО')['Оцека по 5'].mean().sort_values()

plt.figure(figsize=(12, 6))

sns.barplot(x=grade, y=grade.index, palette='coolwarm', edgecolor='black')

plt.title('Средние оценки по видам образовательных учреждений', fontsize=16, fontweight='bold')
plt.xlabel('Средняя оценка по 5-балльной шкале', fontsize=12)
plt.ylabel('Виды образовательных учреждений', fontsize=12)

plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

# Удаление лишних рамок
sns.despine(left=True, bottom=True)

plt.tight_layout()
plt.show()
```

Мы видим, что наилучшие результаты ОГЭ показали гимназии. За ними следуют лицеи, а затем средние общеобразовательные школы с углубленным изучением отдельных предметов. Давайте рассмотрим эти категории более подробно.


```python
# Исходные данные
lic = oge.loc[oge['Вид ОО'] == 'Лицей']
grade_l = lic['Оцека по 5'].mean()

gim = oge.loc[oge['Вид ОО'] == 'Гимназия']
grade_g = gim['Оцека по 5'].mean()

sred = oge.loc[oge['Вид ОО'] == 'Средняя общеобразовательная школа с углубленным изучением отдельных предметов']
grade_s = sred['Оцека по 5'].mean()

data = {
    'Тип образовательного учреждения': ['Лицей', 'Гимназия', 'СОШ с углубленным изучением отдельных предметов'],
    'Средняя оценка': [grade_l, grade_g, grade_s]
}

df = pd.DataFrame(data)

plt.figure(figsize=(10, 6))
sns.set_style("whitegrid")

sns.barplot(x='Тип образовательного учреждения', y='Средняя оценка', data=df, palette='coolwarm', edgecolor='black')

plt.title('Средние оценки по типам образовательных учреждений', fontsize=16, fontweight='bold')
plt.xlabel('Тип образовательного учреждения', fontsize=12)
plt.ylabel('Средняя оценка по 5-балльной шкале', fontsize=12)

plt.xticks(rotation=45, ha='right', fontsize=10)
plt.yticks(fontsize=10)
sns.despine(left=True, bottom=True)

plt.tight_layout()

plt.show()
```

Хотя разница между образовательными организациями незначительная, она все же присутствует. Давайте рассмотрим возможные причины, которые могли повлиять на такие результаты.

```python
gim = oge.loc[oge['Вид ОО'] == 'Гимназия']
count_gim = gim['Выпускник'].value_counts()
print(f'Количество учеников: {count_gim.sum()}')
grades_gim = gim['Оцека по 5'].value_counts()
print(grades_gim)
```

```python
plt.figure(figsize=(10, 6))
sns.set_style('whitegrid')

# Создание барплота с использованием другой цветовой палитры
sns.barplot(x=grades_gim.index, y=grades_gim.values, palette='Blues_r', edgecolor='black')

# Добавление заголовка и меток осей
plt.title('Оценки, полученные выпускниками гимназии', fontsize=16, fontweight='bold')
plt.xlabel('Оценки', fontsize=12)
plt.ylabel('Количество', fontsize=12)

# Настройка шрифтов и поворота меток
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

# Добавление значений над столбцами
for index, value in enumerate(grades_gim):
    plt.text(index, value + 0.5, str(value), ha='center', va='bottom', fontsize=10, color='black')

# Удаление лишних рамок и добавление эстетики
sns.despine(left=True, bottom=True)

# Настройка плотной компоновки
plt.tight_layout()

# Отображение графика
plt.show()
```

###"Профили" отличников, хорошистов, двоечников

"Профиль" отличника по ЕГЭ

```python
ege_top_performers = ege[ege['Балл'] >= 85]
ege_top_performers_task_success_rate = ege_top_performers.apply(lambda row: row[f'Task{row.index}'] / ege_top_performers['Балл'].sum() * 100, axis=1)
ege_top_performers_task_success_rate.sort_values(ascending=False, inplace=True)
print('Профиль отличника по ЕГЭ:')
print(ege_top_performers_task_success_rate)
```

"Профиль" хорошиста по ЕГЭ

```python
ege_good_performers = ege[(ege['Балл'] >= 65) & (ege['Балл'] < 85)]
ege_good_performers_task_success_rate = ege_good_performers.apply(lambda row: row[f'Task{row.index}'] / ege_good_performers['Балл'].sum() * 100, axis=1)
ege_good_performers_task_success_rate.sort_values(ascending=False, inplace=True)
print('Профиль хорошиста по ЕГЭ:')
print(ege_good_performers_task_success_rate)
```

"Профиль" двоечника по ЕГЭ

```python
ege_bad_performers = ege[ege['Балл'] < 40]
ege_bad_performers_task_success_rate = ege_bad_performers.apply(lambda row: row[f'Task{row.index}'] / ege_bad_performers['Балл'].sum() * 100, axis=1)
ege_bad_performers_task_success_rate.sort_values(ascending=False, inplace=True)
print('Профиль двоечника по ЕГЭ:')
print(ege_bad_performers_task_success_rate)
```

"Профиль" отличника по ОГЭ

```python
oge_top_performers = oge[oge['% Оценка'] >= 85]
oge_top_performers_task_success_rate = oge_top_performers.apply(lambda row: row[f'{row.index}_Часть В'] / oge_top_performers['% Оценка'].sum() * 100, axis=1)
oge_top_performers_task_success_rate.sort_values(ascending=False, inplace=True)
print('Профиль отличника по ОГЭ:')
print(oge_top_performers_task_success_rate)
```

"Профиль" хорошиста по ОГЭ

```python
oge_good_performers = oge[(oge['% Оценка'] >= 65) & (oge['% Оценка'] < 85)]
oge_good_performers_task_success_rate = oge_good_performers.apply(lambda row: row[f'{row.index}_Часть В'] / oge_good_performers['% Оценка'].sum() * 100, axis=1)
oge_good_performers_task_success_rate.sort_values(ascending=False, inplace=True)
print('Профиль хорошиста по ОГЭ:')
print(oge_good_performers_task_success_rate)
```

"Профиль" двоечника по ОГЭ

```python
oge_bad_performers = oge[oge['% Оценка'] < 40]
oge_bad_performers_task_success_rate = oge_bad_performers.apply(lambda row: row[f'{row.index}_Часть В'] / oge_bad_performers['% Оценка'].sum() * 100, axis=1)
oge_bad_performers_task_success_rate.sort_values(ascending=False, inplace=True)
print('Профиль двоечника по ОГЭ:')
print(oge_bad_performers_task_success_rate)
```
