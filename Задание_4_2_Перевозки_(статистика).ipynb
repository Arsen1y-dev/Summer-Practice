Возьмите данные о перевозках между посёлками.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import drive
```

```python
file = pd.read_csv('/content/drive/MyDrive/Файлы для pandas/task19.csv', sep=';', encoding='cp1251')
file
```

```python
avg_distance_by_direction = file.groupby('Пункт назначения')['Расстояние'].mean()

plt.figure(figsize=(10, 7))
avg_distance_by_direction.plot(kind='bar')
plt.title('Среднее расстояние по направлениям')
plt.ylabel('Расстояние')
plt.show()
```

```python
plt.figure(figsize=(10, 7))
plt.scatter(file['Расстояние'], file['Расход бензина'])
plt.xlabel('Расстояние')
plt.ylabel('Расход бензина')
plt.title('Расход бензина в зависимости от расстояния')
plt.show()
```

```python
plt.figure(figsize=(10, 7))
sns.boxplot(x='Пункт назначения', y='Расход бензина', data=file)
plt.xlabel('Пункт назначения')
plt.ylabel('Расход бензина')
plt.title('Расход бензина по направлениям')
plt.show()
```
